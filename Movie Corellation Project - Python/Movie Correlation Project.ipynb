```python
#Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjusting config of plots to be created
```

```python
#Read Data
df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
#Check for missing data

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df=df.dropna()
```

```python
#Column data types

df.dtypes
```

```python
#Changing column datatypes

df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df
```

```python
#Correct year
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 20)
```

```python
#Drop Duplicates if any

df.drop_duplicates()
```

```python
#Scatter plot with Budget v/s Gross rev

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget v/s Gross Revenue')
plt.xlabel('Gross Revenue')
plt.ylabel('Film Budget')
plt.show()
```

```python
df.head()
```

```python
#Plotting Budget v/s Gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#Looking at correlation
df.corr()
```

```python
correlation_matrix=df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Converting String columns into numbers

df_num = df

for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes

df_num
        
```

```python
df
```

```python
correlation_matrix_=df_num.corr()

sns.heatmap(correlation_matrix_, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_num.corr()
```

```python
corr_pairs = correlation_matrix_.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
